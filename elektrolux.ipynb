```python
import json

import numpy as np
import pandas as pd
import requests

```

```python
#Download data from elektrolux
url = "https://api.eluxmkt.com/external/commerce/ccv2/occ/POL-ELX/products/compare?query=:relevance:category:kitchen%2fcooking%2fovens%2fsteam-oven&currentPage=0&pageSize=100"

payload = {}
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:132.0) Gecko/20100101 Firefox/132.0',
    'TE': 'trailers',
    'Referer': 'https://www.electrolux.pl/',
    'Origin': 'https://www.electrolux.pl/',
}

response = requests.request("GET", url, headers=headers, data=payload)

file = open("response_elektrolux.json", "w")
file.write(response.text)
file.close()
```

```python
file = open('response_elektrolux.json', 'r')
data = json.load(file)
file.close()
```

```python
for item in data['products']:
    print(item['name'], item['modelId'])
```

```python
products = [{'name': item['name'], 'modelId': item['modelId'], 'specifications': item['specifications']} for item in
            data['products']]
df = pd.DataFrame(products)
df.head()
```

```python
index_main = []
index_sub = []

for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] == 'CookingFunctions':
            for item in spec['value'].split('|'):
                val = 'Tryb_' + item
                if val not in index_sub:
                    index_main.append('Performance')
                    index_sub.append(val)

for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] != 'CookingFunctions':
            if spec['translatedName'] not in index_sub:
                index_main.append(spec['groupName'])
                index_sub.append(spec['translatedName'])

arrays = [
    np.array(index_main),
    np.array(index_sub)
]
arrays
```

```python
model_products = []
for item in data['products']:
    model_products.append(item['modelId'])
model_products = list(set(model_products))
model_products
```

```python
df = pd.DataFrame('---', index=arrays, columns=model_products, dtype='str')
# df.loc[('Performance', 'Tryb_Au Gratin'), 'KODDP77H'] = 'TAK'
df
```

```python
for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] == 'CookingFunctions':
            for s in spec['value'].split('|'):
                val = 'Tryb_' + s
                df.loc[('Performance', val), item['modelId']] = "TAK"
df.head()
```

```python
for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] != 'CookingFunctions':
            df.loc[(spec['groupName'], spec['translatedName']), item['modelId']] = spec['value']
df.head()
```
